```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, bode

# Symbolic setup
s = sp.symbols('s')
Vn, Vout = sp.symbols('Vn Vout')

# Parameters
A = 200000 / (0.04 * s + 1)  # Op-amp transfer function
Vin = 1
R1, R2, C = 1000, 1000, 1e-6

# System equations
eq1 = sp.Eq((Vn - Vin)/(R1 + 1/(s*C)) + (Vn - Vout)/R2, 0)
eq2 = sp.Eq(Vout, A * (-Vn))

# Solve system
solution = sp.solve([eq1, eq2], (Vn, Vout))
Vout_over_Vin = sp.simplify(solution[Vout]/Vin)

# Get coefficients
num, den = sp.fraction(Vout_over_Vin)
num_coeffs = [float(c) for c in sp.Poly(num, s).all_coeffs()]
den_coeffs = [float(c) for c in sp.Poly(den, s).all_coeffs()]

print("Transfer Function:", Vout_over_Vin)
print(f"Numerator: {num_coeffs}")
print(f"Denominator: {den_coeffs}")

# Frequency analysis
omega = np.logspace(0, 8, 1000)  # Frequency range
sys = TransferFunction(num_coeffs, den_coeffs)
w, mag, phase = bode(sys, omega)

# Plotting
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.semilogx(w, mag)
plt.title('Bode Plot of LM741 Op-Amp Circuit')
plt.ylabel('Magnitude (dB)')
plt.grid(which='both', alpha=0.5)

plt.subplot(2, 1, 2)
plt.semilogx(w, phase, color='orange')
plt.xlabel('Frequency (rad/s)')
plt.ylabel('Phase (degrees)' )
plt.grid(which='both', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Given values
R1 = 1000  # 1k ohm
R2 = 1000  # 1k ohm
C = 1e-6   # 1uF

# Coefficients
num = [-R2*C, 0]      # numerator coefficients
den = [R1*C, 1]       # denominator coefficients

# Create transfer function
system = signal.TransferFunction(num, den)

# Define custom frequency range
w = np.linspace(0.01, 10000, 100)  

# Get magnitude and phase
w, mag, phase = signal.bode(system, w=w)

# Plot magnitude and phase as separate subplots
plt.figure(figsize=(12, 8))

# Magnitude plot
plt.subplot(2, 1, 1)
plt.semilogx(w, mag, label='Magnitude [dB]')
plt.title('Bode Plot of Ideal Op-Amp Circuit')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Magnitude [dB]')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

# Phase plot
plt.subplot(2, 1, 2)
plt.semilogx(w, phase, label='Phase [degrees]', color='orange')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Phase [degrees]')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import sympy as sp

# Common setup
s = sp.symbols('s')
Vn, Vout = sp.symbols('Vn Vout')
Vin = 1
R1, R2, C = 1000, 1000, 1e-6
w = np.logspace(2, 7, 1000)  # 100 to 1M rad/s, tighter range

# Define transfer function calculator
def get_tf(A_expr):
    eq1 = sp.Eq((Vn - Vin)/(R1 + 1/(s*C)) + (Vn - Vout)/R2, 0)
    eq2 = sp.Eq(Vout, A_expr * (-Vn))
    solution = sp.solve([eq1, eq2], (Vn, Vout))
    Vout_over_Vin = sp.simplify(solution[Vout]/Vin)
    num, den = sp.fraction(Vout_over_Vin)
    return [float(c) for c in sp.Poly(num, s).all_coeffs()], [float(c) for c in sp.Poly(den, s).all_coeffs()]

# Define all amplifier models

num_lm741, den_lm741 = get_tf(2e4 / (0.04 * s + 1))     # LM741, adjusted
num_op27, den_op27 = get_tf(2e5 / (0.0318 * s + 1))     # OP27, adjusted
num_op77, den_op77 = get_tf(2e6 / (1.59 * s + 1))       # OP77, adjusted
num_ideal = [-R2*C, 0]
den_ideal = [R1*C, 1]

# Plot
plt.figure(figsize=(14, 10))

# Magnitude plot
plt.subplot(2, 1, 1)
for name, num, den, color, linestyle in [
    ('Ideal', num_ideal, den_ideal, 'black', '-'),
    ('LM741', num_lm741, den_lm741, 'blue', '--'),
    ('OP27', num_op27, den_op27, 'green', '-.'),
    ('OP77', num_op77, den_op77, 'red', ':')
]:
    _, mag, _ = signal.bode((num, den), w)
    plt.semilogx(w, mag, label=name, color=color, linestyle=linestyle, linewidth=2)

plt.title('Comparative Bode Plot of High-Pass Filters')
plt.ylabel('Magnitude (dB)')
plt.grid(which='both', linestyle='--', alpha=0.7)
plt.legend()
plt.ylim(-20, 5)  # Tighten magnitude range

# Phase plot
plt.subplot(2, 1, 2)
for name, num, den, color, linestyle in [
    ('Ideal', num_ideal, den_ideal, 'black', '-'),
    ('LM741', num_lm741, den_lm741, 'blue', '--'),
    ('OP27', num_op27, den_op27, 'green', '-.'),
    ('OP77', num_op77, den_op77, 'red', ':')
]:
    _, _, phase = signal.bode((num, den), w)
    plt.semilogx(w, phase, label=name, color=color, linestyle=linestyle, linewidth=2)

plt.xlabel('Frequency (rad/s)')
plt.ylabel('Phase (degrees)')
plt.grid(which='both', linestyle='--', alpha=0.7)
plt.yticks(np.arange(-180, 181, 45))
plt.ylim(-280,-90)
plt.legend()

plt.tight_layout()
plt.show()
```
